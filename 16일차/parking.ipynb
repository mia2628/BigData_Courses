```python
import pandas as pd
```

```python
주차장현황 = pd.read_csv('제주도_주차장현황.csv', encoding = 'euc_kr')
주차장현황
```

```python
a=pd.DataFrame(주차장현황.iloc[6:,-2:])
a
```

```python
b = pd.DataFrame(주차장현황.iloc[6:,0:2])
b
```

```python
c = pd.concat([b,a], axis = 1)
d= c.iloc[:-6,]
d
```

```python
e=d[d['행정구역(시)별']!='합계']
e.dropna(axis = 0, inplace = True)
e
```

```python
십칠=e.iloc[:,:3]
십칠['2017 년'] = 십칠['2017 년'].astype(int)
십칠
```

```python
십칠2 = 십칠.groupby(['주차장종류별', '행정구역(시)별']).sum().reset_index(drop=False)
십칠2
```

```python
십팔=e.drop(['2017 년'], axis = 1)
십팔['2018 년'] = 십팔['2018 년'].astype(int)
십팔
```

```python
십팔2 = 십팔.groupby(['주차장종류별', '행정구역(시)별']).sum().reset_index(drop=False)
십팔2
```
