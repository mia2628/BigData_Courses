```python
import pandas as pd
import numpy as np
from haversine import haversine

distance = pd.read_csv("C:\\Users\\USER\\Desktop\\샘플데이터\\정제 데이터\\전처리데이터.csv", engine='python')
```

```python
distance.fillna(0, inplace=True)
```

```python
distance
```

```python
distance['거리_전기차'] = 0
distance['거리_충전소'] = 0
distance['거리_교통량'] = 0
```

```python
distance
```

```python
distance2=distance[distance['전기차']!=0]
```

```python
for j in range(0, len(distance)):
    for i in range(0, len(distance2)):
        lon = distance['경도'][j]
        lat = distance['위도'][j]
        lon2 = distance2['경도'].reset_index(drop=True)[i]
        lat2 = distance2['위도'].reset_index(drop=True)[i]
        a = (lat, lon)
        b = (lat2, lon2)
        c = haversine(a, b) * 1000
        if c <= 500:
             distance['거리_전기차'][j] = 1
```

```python
distance3=distance[distance['충전소']!=0]
```

```python
for j in range(0, len(distance)):
    for i in range(0, len(distance3)):
        lon = distance['경도'][j]
        lat = distance['위도'][j]
        lon2 = distance3['경도'].reset_index(drop=True)[i]
        lat2 = distance3['위도'].reset_index(drop=True)[i]
        a = (lat, lon)
        b = (lat2, lon2)
        c = haversine(a, b) * 1000
        if c <= 500:
             distance['거리_충전소'][j] = 1
```

```python
distance4=distance[distance['교통량']!=0]
```

```python
for j in range(0, len(distance)):
    for i in range(0, len(distance4)):
        lon = distance['경도'][j]
        lat = distance['위도'][j]
        lon2 = distance4['경도'].reset_index(drop=True)[i]
        lat2 = distance4['위도'].reset_index(drop=True)[i]
        a = (lat, lon)
        b = (lat2, lon2)
        c = haversine(a, b) * 1000
        if c <= 500:
             distance['거리_교통량'][j] = 1
```

```python
distance.to_csv("C:\\Users\\USER\\Desktop\\샘플데이터\\정제 데이터\\전처리데이터2.csv",
                header = True, index=False, sep=',', encoding='cp949')
```
